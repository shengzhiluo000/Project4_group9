```python
from pathlib import Path
import pandas as pd
```

```python
data = Path('Resource/Diabetes_Prediction_2018_2022.csv')
df = pd.read_csv(data)
df.head()
```

```python
df["Diabetes_012"] = df["Diabetes_012"].replace(2,1)
df["Diabetes_012"].head()
df1 = df[(df["Diabetes_012"]==0)]
df2 = df[(df["Diabetes_012"]==1)]
#df["Diabetes_012"].value_counts()
df1
```

```python
df2["Diabetes_012"].value_counts()
```

```python
no_to_drop = len(df1["Diabetes_012"])-len(df2["Diabetes_012"])
df1 = df1.drop(df1.index[:no_to_drop])
df1.reset_index(inplace=True,drop=True)
```

```python
df1
```

```python
df = pd.concat([df1,df2], axis=0, ignore_index= True)
df
```

```python
from sklearn.preprocessing import StandardScaler
df_scaled = df.drop(columns=["Diabetes_012"])

diabete_scaled = StandardScaler().fit_transform(df_scaled)
```

```python
df_scaled = pd.DataFrame(diabete_scaled)
```

```python
y = df["Diabetes_012"]
X = df_scaled
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    random_state=1, 
                                                    stratify=y)
X_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver='lbfgs',
                                max_iter=1000,
                                random_state=1)
classifier
```

```python
classifier.fit(X_train, y_train)
```

```python
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

```python
predictions = classifier.predict(X_test)
results = pd.DataFrame({"Prediction": predictions, "Actual": y_test}).reset_index(drop=True)
results.head(100)
```

```python
results["Actual"].value_counts()
```

```python
results["Prediction"].value_counts()
```

```python
from sklearn.metrics import accuracy_score
# Display the accuracy score for the test dataset.
accuracy_score(y_test, predictions)
```
